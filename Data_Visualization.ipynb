How does the child mortality rate in the country affect the number of children per woman in this country?  

- The countries with higher child mortality rates tend to have more children per woman. 

### Imports and Raw Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
babyPerWoman_data = pd.read_csv('https://raw.githubusercontent.com/Catmint-V/Data-Visualization/master/data/children_per_woman_total_fertility.csv')

childmortality_data = pd.read_csv('https://raw.githubusercontent.com/Catmint-V/Data-Visualization/master/data/child_mortality_0_5_year_olds_dying_per_1000_born.csv')

income_data = pd.read_csv('https://raw.githubusercontent.com/Catmint-V/Data-Visualization/master/data/income_per_person_gdppercapita_ppp_inflation_adjusted.csv')

geo_data = pd.read_csv('https://raw.githubusercontent.com/ef2020/TextFiles/master/gap_minder_geography.csv')
```

```python
df = pd.DataFrame(babyPerWoman_data)
df
```

```python
babyPerWoman = pd.DataFrame(df.mean().reset_index())
babyPerWoman.columns = ['Year', 'ChildPerWoman']
babyPerWoman
```

```python
fig, ax = plt.subplots(figsize=(20,7))

ax.plot(babyPerWoman['Year'], babyPerWoman['ChildPerWoman'])
ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))

ax.set_xlabel('Year')
ax.set_ylabel('Child Per Woman')
plt.title('Average Number of Child Per Woman in the World', fontdict={'fontsize': 25})

plt.show()
```

```python
childmortality = pd.DataFrame(childmortality_data)
childmortality
```

```python
income = pd.DataFrame(income_data)
income
```

```python
geo = pd.DataFrame(geo_data)
geo
```

### Data Cleansing

- Keep data only from year 2020
- merge 3 dataframe into 1

```python
df = df.loc[:, ['country', '2020']]
childmortality = childmortality.loc[:, ['country', '2020']]
income = income.loc[:, ['country', '2020']]
```

```python
geo = geo.loc[:, ['geo', 'name', 'four_regions', 'eight_regions', 'six_regions', 'Latitude', 'Longitude', 'UNHCR']]
```

##### Merge Dataframe into One 

```python
result = pd.merge(df, childmortality, on='country', how='outer')
result = pd.merge(result, income, on='country', how='outer')
result.columns = ['Country', 'ChildPerWoman', 'ChildMortality', 'Income']

result
```

##### Check Null Values

- All null values are dropped

```python
result.isnull().sum()
```

```python
result_null = result[result.isna().any(axis=1)]
result_null
```

```python
result = result.dropna()
```

```python
result
```

##### Change Data Type

```python
result.dtypes
```

```python
result['Income'] = result['Income'].apply(lambda x: float(x.replace('k', '')) * 1000 if 'k' in x else float(x))
```

##### Merge result dataframe with geo dataframe

```python
result = pd.merge(result, geo, how='left', left_on='Country', right_on='name')
result = result.drop('name', axis=1)
result
```

### Data Visualization

```python
# demonstrate the connection between the income and number of children (x and y axis)
# the size of the bubble corresponds to the child mortality rate
# the color of the bubble corresponds to the value of the four-regions attribute from the geo dataframe

plt.rcParams["figure.figsize"] = (12,9)

sns.scatterplot(data=result, x='Income', y='ChildPerWoman', size='ChildMortality', hue='four_regions', sizes=(20, 2000), alpha=0.8)

plt.title('Income V.S. Number of Child Per Woman', fontdict={'fontsize': 25})
plt.xlabel('Income')
plt.ylabel('Child Per Woman')

plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0, handlelength=5, handleheight=6, fontsize = 10)

sns.set_style('whitegrid')
sns.set(font_scale=1.2, rc={'figure.figsize':(12,9), 'grid.alpha': 0.7, 'legend.markerscale': 1})
# plt.grid()

plt.show()
```

### Conclusions

- The higher the income, the less child per woman.
- The higher the child mortality, the higher the child per woman.
- The higher the child mortality, the lower the income.
- Africa region has the most child mortality.
- Europe region has the least child mortality.
- Africa region has more child per woman.
- Europe and Americas region has less child per woman.
